```
# point geometry 생성 
## 위도 경도를 geometry로 변경해 준다.
def to_float(df):
    df['geometry'] = df.apply(lambda row : Point([row['경도'], row['위도']]),axis=1)
    df = gpd.GeoDataFrame(df, geometry='geometry')
    return df
```

### haversine 직선거리 구해주기

```
# haversine 구하는 식
station_1 = []
station_2 = []
harv = []
for sta, lat, long in zip(Bike_station['Station_ID'],Bike_station['위도'],Bike_station['경도']):
    for sta1, lat1, long1 in zip(Bike_station['Station_ID'],Bike_station['위도'],Bike_station['경도']):
        x = (lat, long)
        y = (lat1, long1) 
        h = haversine(x, y) *1000
        station_1.append(sta)
        station_2.append(sta1)
        harv.append(h)
```

```
# 스테이션 별 이동거리 dataframe 만들어 주기
df_harv = pd.DataFrame(data={'Station_LEAS':station_1, 'Station_RTN':station_2, 'harv(m)':harv})
df_harv['Station_LEAS'] = df_harv[['Station_LEAS']].astype(str)
df_harv['Station_RTN'] = df_harv[['Station_RTN']].astype(str)
df_harv['LEAS_RTN'] = df_harv[['Station_LEAS','Station_RTN']].apply(lambda x:'_'.join(x), axis=1) # 편하게 보기위해 LEAS와 RTN을 합친다
df_harv
```
