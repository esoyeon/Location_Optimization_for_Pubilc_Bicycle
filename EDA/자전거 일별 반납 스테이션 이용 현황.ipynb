### 대여 스테이션 이용 횟수를 기준으로 

- 대여시간 별로 각각의 반납 스테이션 이용 횟수
- pivot data
- max

- 4분위(25)
- 4분위(75)

```
# group by를 통해 대여시간 별로 각각의 반납 스테이션 이용 횟수
station_day_return_count = (df_OH_5.groupby([df_OH_5['대여 시간'],df_OH_5['반납 스테이션 번호']])).count()
df_station_day_return_count = station_day_return_count[['반납 시간']]
df_station_day_return_count.columns = ['스테이션 이용 횟수']
df_station_day_return_count_1 = df_station_day_return_count.reset_index()
df_station_day_return_count_1
```

```
# pivot 데이터를 활용해서 최소,최대,평균 반납 스테이션과 이용 횟수 구하기
df_station_day_return_count_pivot = df_station_day_return_count_1.pivot('반납 스테이션 번호','대여 시간','스테이션 이용 횟수')
df_station_day_return_count_pivot_1 = df_station_day_return_count_pivot.fillna(0) # 0값으로 채워준다.
df_station_day_return_count_pivot_1
```

```
# 스테이션 별 반납스테이션 최대 이용 횟수 및 최대 이용 날짜
def top(df):
    return df.sort_values(by="스테이션 이용 횟수", ascending=False)[:1]

df_station_day_return_maxcount_1 = df_station_day_return_count_1.groupby(df_station_day_return_count_1['반납 스테이션 번호']).apply(top)
df_station_day_return_maxcount_2 = df_station_day_return_maxcount_1.drop(['반납 스테이션 번호'],axis=1)
df_station_day_return_maxcount_3 = df_station_day_return_maxcount_2.reset_index()
df_station_day_return_maxcount_end = df_station_day_return_maxcount_3.drop(['level_1'],axis=1)
df_station_day_return_maxcount_end.columns = ['반납 스테이션 번호','최대 이용일', '최대 스테이션 이용 횟수']
df_station_day_return_maxcount_end
```

```
# 스테이션 별 반납스테이션 최소 이용 횟수 및 최소 이용 날짜
def bottom(df):
    return df.sort_values(by="스테이션 이용 횟수", ascending=True)[:1]

df_station_day_return_mincount_1 = df_station_day_return_count_1.groupby(df_station_day_return_count_1['반납 스테이션 번호']).apply(bottom)
df_station_day_return_mincount_2 = df_station_day_return_mincount_1.drop(['반납 스테이션 번호'], axis=1)
df_station_day_return_mincount_3 = df_station_day_return_mincount_2.reset_index()
df_station_day_return_mincount_end = df_station_day_return_mincount_3.drop(['level_1'], axis=1)
df_station_day_return_mincount_end.columns = ['반납 스테이션 번호','최소 이용일', '최소 스테이션 이용 횟수']
df_station_day_return_mincount_end
```

```
# 스테이션 별 median()중간수
def median(df):
    return df.median()

df_station_day_return_mediancount_1 = df_station_day_return_count_1.groupby(df_station_day_return_count_1['반납 스테이션 번호']).apply(median)
df_station_day_return_mediancount_2 = df_station_day_return_mediancount_1.drop(['반납 스테이션 번호'], axis=1)
df_station_day_return_mediancount_end = df_station_day_return_mediancount_2.reset_index()
df_station_day_return_mediancount_end.columns = ['반납 스테이션 번호','스테이션 이용 횟수(quartile50)']
df_station_day_return_mediancount_end
```

```
# 스테이션 별 4분위-25
def quantile_25(df):
    return df.quantile(.25)

df_station_day_return_qua25count_1 = df_station_day_return_count_1.groupby(df_station_day_return_count_1['반납 스테이션 번호']).apply(quantile_25)
df_station_day_return_qua25count_end = df_station_day_return_qua25count_1.reset_index()
df_station_day_return_qua25count_end.columns = ['반납 스테이션 번호','스테이션 이용 횟수(quartile25)']
df_station_day_return_qua25count_end
```

```
# 스테이션 별 4분위-75
def quantile_75(df):
    return df.quantile(.75)

df_station_day_return_qua75count_1 = df_station_day_return_count_1.groupby(df_station_day_return_count_1['반납 스테이션 번호']).apply(quantile_75)
df_station_day_return_qua75count_end = df_station_day_return_qua75count_1.reset_index()
df_station_day_return_qua75count_end.columns = ['반납 스테이션 번호','스테이션 이용 횟수(quartile75)']
df_station_day_return_qua75count_end
```

```
# 스테이션 별 평균
def mean(df):
    return df.mean().round(2)
 
df_station_day_return_meancount_1 = df_station_day_return_count_1.groupby(df_station_day_return_count_1['반납 스테이션 번호']).apply(mean) 
df_station_day_return_meancount_2 = df_station_day_return_meancount_1[['스테이션 이용 횟수']]
df_station_day_return_meancount_end = df_station_day_return_meancount_2.reset_index()
df_station_day_return_meancount_end.columns = ['반납 스테이션 번호','스테이션 평균 이용 횟수']
df_station_day_return_meancount_end
```

### 스테이션 별  평균 이용 시간 / 평균 추정 이동 거리 

```
## 평균 이용 시간 구하기
df_OH_return_mean_time = df_OH_5[['대여 시간','반납 스테이션 번호','이용 시간(s)']] # 필요한 컬럼만 가져 온다.
df_OH_return_mean_time_1 = (df_OH_return_mean_time.groupby([df_OH_return_mean_time['대여 시간'],df_OH_return_mean_time['반납 스테이션 번호']])).mean().round(2)
df_OH_return_mean_time_2 = df_OH_return_mean_time_1.reset_index() # index기준을 변경해준다
df_OH_return_mean_time_2
```

```
def mean(df):
    return df.mean().round(0)

df_OH_return_mean_time_3 = df_OH_return_mean_time_2.groupby(df_OH_return_mean_time_2['반납 스테이션 번호']).apply(mean)
df_OH_return_mean_time_3
```

```
df_OH_return_mean_time_4 = df_OH_return_mean_time_3[['이용 시간(s)']]
df_OH_return_mean_time_5 = df_OH_return_mean_time_4.reset_index()
df_OH_return_mean_time_5.columns = ['반납 스테이션 번호','평균 이용 시간(s)']
df_OH_return_mean_time_end = df_OH_return_mean_time_5.assign(TEST=lambda x: ((x['평균 이용 시간(s)']/60).round(1)))
df_OH_return_mean_time_end.columns =['반납 스테이션 번호','평균 이용 시간(s)','평균 이용 시간(m)']
df_OH_return_mean_time_end
```

```
df_OH_return_est_dis = df_OH_5[['대여 시간','반납 스테이션 번호','추정 이동 거리']] # 필요한 컬럼만 뽑아준다.
df_OH_return_est_dis_1 = df_OH_return_est_dis.dropna(axis=0) # 추정 이동 거리 가 있는 평균을 알고 싶기에 처음 NaN 값을 제거해준다.
df_OH_return_est_dis_2 = df_OH_return_est_dis_1[df_OH_return_est_dis_1['추정 이동 거리']!=0] # 0으로 된 추정이동거리 값도 제거해준다.
df_OH_return_est_dis_2
```
