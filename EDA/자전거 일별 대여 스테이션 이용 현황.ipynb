### 대여 스테이션 이용 횟수를 기준으로

```python
# group by를 통해 대여시간 별로 각각의 대여 스테이션 이용 횟수
station_day_rent_count = (df_OH_5.groupby([df_OH_5['대여 시간'],df_OH_5['대여 스테이션 번호']])).count()
df_station_day_rent_count = station_day_rent_count[['대여 STATION_NAME']]
df_station_day_rent_count.columns = ['스테이션 이용 횟수']
df_station_day_rent_count
```

```python
# pivot data를 만들기 위한 과정 
df_station_day_rent_count_1 = df_station_day_rent_count.reset_index()
df_station_day_rent_count_1
```

```python
# pivot 데이터를 활용해서 최소,최대,평균 대여 스테이션과 이용 횟수 구하기
df_station_day_rent_count_pivot = df_station_day_rent_count_1.pivot('대여 스테이션 번호','대여 시간','스테이션 이용 횟수')
df_station_day_rent_count_pivot_1 = df_station_day_rent_count_pivot.fillna(0) # 0값으로 채워준다.
df_station_day_rent_count_pivot_1
```

```python
# 스테이션별 대여 스테이션 최대 이용 횟수 및 최대 이용 날짜
def top(df):
    return df.sort_values(by="스테이션 이용 횟수", ascending=False)[:1]

df_station_day_rent_maxcount_2=  df_station_day_rent_count_1.groupby(df_station_day_rent_count_1['대여 스테이션 번호']).apply(top)
df_station_day_rent_maxcount_3 = df_station_day_rent_maxcount_2[['대여 시간','스테이션 이용 횟수']]
df_station_day_rent_maxcount_3.columns = ['최대 이용일', '최대 스테이션 이용 횟수']
df_station_day_rent_maxcount_4 = df_station_day_rent_maxcount_3.reset_index()
df_station_day_rent_maxcount_end = df_station_day_rent_maxcount_4.drop(['level_1'], axis=1)
df_station_day_rent_maxcount_end
 
```

```python
# 스테이션 별 대여 스테이션 최소 이용 횟수 및 최소 이용 날짜 
def bottom(df):
    return df.sort_values(by="스테이션 이용 횟수", ascending=True)[:1]

df_station_day_rent_mincount_2 = df_station_day_rent_count_1.groupby(df_station_day_rent_count_1['대여 스테이션 번호']).apply(bottom)
df_station_day_rent_mincount_3 = df_station_day_rent_mincount_2.drop(['대여 스테이션 번호'], axis=1)
df_station_day_rent_mincount_4 = df_station_day_rent_mincount_3.reset_index()
df_station_day_rent_mincount_end = df_station_day_rent_mincount_4.drop(['level_1'], axis=1)
df_station_day_rent_mincount_end.columns = ['대여 스테이션 번호','최소 이용일','최소 스테이션 이용 횟수']
df_station_day_rent_mincount_end
```

```python
# 스테이션 별 median()중간수
def median(df):
    return df.median()

df_station_day_rent_mediancount_1 = df_station_day_rent_count_1.groupby(df_station_day_rent_count_1['대여 스테이션 번호']).apply(median)
df_station_day_rent_mediancount_2 = df_station_day_rent_mediancount_1.drop(['대여 스테이션 번호'], axis=1)
df_station_day_rent_mediancount_end = df_station_day_rent_mediancount_2.reset_index()
df_station_day_rent_mediancount_end.columns = ['대여 스테이션 번호', '스테이션 이용 횟수(quartile50)']
df_station_day_rent_mediancount_end
```

```python
# 스테이션 별 1/분위 값
def quantile_25(df):
    return df.quantile(.25)

df_station_day_rent_qua25count_1 = df_station_day_rent_count_1.groupby(df_station_day_rent_count_1['대여 스테이션 번호']).apply(quantile_25)
df_station_day_rent_qua25count_end = df_station_day_rent_qua25count_1.reset_index()
df_station_day_rent_qua25count_end.columns = ['대여 스테이션 번호', '스테이션 이용 횟수(quartile25)']
df_station_day_rent_qua25count_end
```

```python
# 스테이션 별 4분위-75
def quantile_75(df):
    return df.quantile(.75)

df_station_day_rent_qua75count_1 = df_station_day_rent_count_1.groupby(df_station_day_rent_count_1['대여 스테이션 번호']).apply(quantile_75)
df_station_day_rent_qua75count_end = df_station_day_rent_qua75count_1.reset_index()
df_station_day_rent_qua75count_end.columns = ['대여 스테이션 번호', '스테이션 이용 횟수(quartile75)']
df_station_day_rent_qua75count_end
```

```python
# 스테이션 별 평균
def mean(df):
    return df.mean().round(2)

f_station_day_rent_meancount = df_station_day_rent_count_1.groupby(df_station_day_rent_count_1['대여 스테이션 번호']).apply(mean)
f_station_day_rent_meancount_1 = f_station_day_rent_meancount.drop(['대여 스테이션 번호'], axis=1)
f_station_day_rent_meancount_end = f_station_day_rent_meancount_1.reset_index()
f_station_day_rent_meancount_end.columns = ['대여 스테이션 번호', '스테이션 평균 이용 횟수']
f_station_day_rent_meancount_end
```

### 스테이션 별 평균 추정 이동 거리 / 평균 이용 시간

```python
df_OH_est_dis = df_OH_5[['대여 시간','대여 스테이션 번호','추정 이동 거리']] # 필요한 컬럼만 뽑아준다.
df_OH_est_dis_1 = df_OH_est_dis.dropna(axis=0) # 추정 이동 거리 가 있는 평균을 알고 싶기에 처음 NaN 값을 제거해준다.
df_OH_est_dis_2 = df_OH_est_dis_1[df_OH_est_dis_1['추정 이동 거리']!=0] # 0으로 된 추정이동거리 값도 제거해준다.
df_OH_est_dis_2
```

```python
df_est_dis_mean = (df_OH_est_dis_2.groupby([df_OH_est_dis_2['대여 시간'],df_OH_est_dis_2['대여 스테이션 번호']])).mean()
df_est_dis_mean_1 = df_est_dis_mean.reset_index()# index기준을 변경해준다.
df_est_dis_mean_1
```

```python
# 스테이션을 기준으로 3년간 추정 이동 거리의 평균을 구해 준다.
def mean(df):
    return df.mean().round(2)

df_est_dis_mean_1 = df_est_dis_mean_1.groupby(df_est_dis_mean_1['대여 스테이션 번호']).apply(mean)
df_est_dis_mean_1
```

```python
df_est_dis_mean_1.columns = ['1','추정 이동 거리']
df_est_dis_mean_2 = df_est_dis_mean_1.reset_index()
df_est_dis_mean_end = df_est_dis_mean_2[['대여 스테이션 번호','추정 이동 거리']]
df_est_dis_mean_end.columns = ['대여 스테이션 번호','추정 이동 거리 평균']
df_est_dis_mean_end
```
