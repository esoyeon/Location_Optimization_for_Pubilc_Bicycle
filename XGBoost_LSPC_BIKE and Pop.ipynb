```
from tqdm.notebook import tqdm
lscp_pop_and_bike['DN_avg'] = np.zeros(len(lscp_pop_and_bike))
lscp_pop_and_bike['Road_yes'] = np.zeros(len(lscp_pop_and_bike))
lscp_pop_and_bike['Space_count'] = np.zeros(len(lscp_pop_and_bike))
lscp_pop_and_bike['Culture_count'] = np.zeros(len(lscp_pop_and_bike))
lscp_pop_and_bike['Sport_count'] = np.zeros(len(lscp_pop_and_bike))
lscp_pop_and_bike['KINTEX'] = np.zeros(len(lscp_pop_and_bike))
lscp_pop_and_bike['Hosu_yes'] = np.zeros(len(lscp_pop_and_bike))
lscp_pop_and_bike['Park_count'] = np.zeros(len(lscp_pop_and_bike))


for i in tqdm(range(len(lscp_pop_and_bike))):
  for j in range(len(lscp_bike_300)):
    if str(lscp_bike_300['Bike_Station'].iloc[j]) in lscp_pop_and_bike['스테이션'].iloc[i]:
      lscp_pop_and_bike['DN_avg'].iloc[i] += lscp_bike_300['DN_avg'].iloc[j]
      lscp_pop_and_bike['Road_yes'].iloc[i] += lscp_bike_300['Road_yes'].iloc[j]
      lscp_pop_and_bike['Space_count'].iloc[i] += lscp_bike_300['Space_count'].iloc[j]
      lscp_pop_and_bike['Culture_count'].iloc[i] += lscp_bike_300['Culture_count'].iloc[j]
      lscp_pop_and_bike['Sport_count'].iloc[i] += lscp_bike_300['Sport_count'].iloc[j]
      lscp_pop_and_bike['KINTEX'].iloc[i] += lscp_bike_300['KINTEX'].iloc[j]
      lscp_pop_and_bike['Hosu_yes'].iloc[i] += lscp_bike_300['Hosu_yes'].iloc[j]
      lscp_pop_and_bike['Park_count'].iloc[i] += lscp_bike_300['Park_count'].iloc[j]
```

# 1. 인구로만 모델링

```
y = lscp_pop_and_bike[['총 이용횟수']]
X = lscp_pop_and_bike[['주거인구', '버스_승하차수', '지하철_승하차수','유동인구']]

X.shape, y.shape
```

```
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X, y, random_state = 42, test_size = 0.1)

X_train.shape, X_test.shape, y_train.shape
```

## 수치형 스케일링

```
numerical_columns = ['주거인구', '버스_승하차수', '지하철_승하차수','유동인구']
```

```
from sklearn.preprocessing import StandardScaler
std_scaler = StandardScaler()
```

```
std_scaler.fit(X_train[numerical_columns])
```

```
std_scaler.transform(X_train[numerical_columns])
std_scaler.transform(X_test[numerical_columns])
```

## 범주형 전처리

## 모델링 

```
import xgboost as xgb

xgb_model = xgb.XGBRegressor(n_estimators=500, 
                              learning_rate = 0.1, 
                              max_depth = 3, objective = 'reg:squarederror').fit(X_train, y_train)
```

```
xgb_model
```

```
print('train score : {}'.format(xgb_model.score(X_train, y_train)))
print('test score : {}'.format(xgb_model.score(X_test, y_test)))
```

```
np.argsort(xgb_model.feature_importances_)
```

```
feature_important = [ X_train.columns[i] for i in np.argsort(xgb_model.feature_importances_)[::-1]]
feature_important
```
